```python
import sys, os
import torch
import numpy as np
import json
import datetime

from dynamics.process.rnn import parse

%load_ext autoreload
%autoreload 2
```

```python
savedir = '/Users/dhocker/projects/kind_cl/data/rnndata_matlab/overtraining/'
datadir = '/Users/dhocker/projects/dynamics/results/20231003/full_cl/'
```

```python
num = 26
#name = 'rnn_'+str(num)+'_allbeh_1k.json'
fname = 'rnn_'+str(num)+'_allbeh.json'
with open(datadir+fname,'rb') as f:
    dat = json.load(f)
    
```

```python
#iterate over numbers
use1k = False

for num in range(1,51):

    
    print(num)
    if use1k:
        fname = 'rnn_'+str(num)+'_allbeh_1k.json'
    else:
        fname = 'rnn_'+str(num)+'_allbeh.json'
    
    with open(datadir+fname,'rb') as f:
        dat = json.load(f)

    #find block sessions 
    idxlist = []
    blocklist = []
    for k in range(len(dat)):
        if dat[k]['name'].find('block') > 0:
            blocklist.append(k)
            if use1k:
                # if 1k.json
                idxlist.append(int(dat[k]['name'].split('_')[-2]))
            else:
                # if 1k.json
                idxlist.append(int(dat[k]['name'].split('_')[-1].strip('.json')))


    blocklist = np.array(blocklist)[np.argsort(np.array(idxlist))]



    # break into similar-sized seddioss, then aggreate back to big session
    datlist_main = []
    tokeep_main = []
    ntrials_main = []

    for j in blocklist:

        datlist, tokeep = parse.format2sess(dat[j], useblocks=True, blockspersess=5)
        ntrials = [len(k['rewvols']) for k in datlist]
        ntrials_2 = [len(k['outcomes']) for k in datlist]
        ntrials_3 = [len(k['blocks']) for k in datlist]
        
        for m in range(len(datlist)):
            datlist_main.append(datlist[m])
            tokeep_main.append(tokeep[m])
            if tokeep[m]:
                ntrials_main.append(ntrials[m])

    #aggregate into large session
    dat_test = parse.format2singlesess(datlist_main, tokeep_main)
    


    # save as matlab file
    if use1k:
        savename = savedir+'rnn_'+str(num)+'_allbeh_1k.mat'
    else:
        savename = savedir+'rnn_'+str(num)+'_allbeh.mat'

    A = parse.generate_rattrial_rnn(dat_test, savename,
                              date_base=str(datetime.date.today())+' ', ops=None, ntrials=ntrials_main)
```
